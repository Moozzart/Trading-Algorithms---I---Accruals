```python
import pandas as pd
import numpy as np
```

```python
company_data = pd.read_excel('Company_data (1).xlsx')
```

```python
x=company_data.dropna()
x.columns
```

```python
x=x[x['L - 6Shares traded']+x['L - 5Shares traded']+x['L - 4Shares traded']+x['L - 3Shares traded']+x['L - 2Shares traded']+x['L - 1Shares traded']>9000]
```

```python
income = pd.concat([x['YEAR _7Total Income'],x['YEAR _6Total income'],x['YEAR _5Total income'],x['YEAR _4Total income'],x['YEAR _3Total income'],x['YEAR _2Total income'],x['YEAR _1Total income']],axis=1)
```

```python
current_assets = pd.concat([x['YEAR _7Current assets (incl. short term investments, loans & advances)'],x['YEAR _6Current assets (incl. short term investments, loans & advances)'],x['YEAR _5Current assets (incl. short term investments, loans & advances)'],x['YEAR _4Current assets (incl. short term investments, loans & advances)'],x['YEAR _3Current assets (incl. short term investments, loans & advances)'],x['YEAR _2Current assets (incl. short term investments, loans & advances)'],x['YEAR _1Current assets (incl. short term investments, loans & advances)']],axis=1)
cash_balance = pd.concat([x['YEAR _7Cash balance'],x['YEAR _6Cash balance'],x['YEAR _5Cash balance'],x['YEAR _4Cash balance'],x['YEAR _3Cash balance'],x['YEAR _2Cash balance'],x['YEAR _1Cash balance']],axis=1)
current_liabilities = pd.concat([x['YEAR _7Current liabilities'],x['YEAR _6Current liabilities'],x['YEAR _5Current liabilities'],x['YEAR _4Current liabilities'],x['YEAR _3Current liabilities'],x['YEAR _2Current liabilities'],x['YEAR _1Current liabilities']],axis=1)
shortterm_brrw = pd.concat([x['YEAR _7Short-term borrowings'],x['YEAR _6Short-term borrowings'],x['YEAR _5Short-term borrowings'],x['YEAR _4Short-term borrowings'],x['YEAR _3Short-term borrowings'],x['YEAR _2Short-term borrowings'],x['YEAR _1Short-term borrowings']],axis=1)
tot_assets = pd.concat([x['YEAR _7Total assets'],x['YEAR _6Total assets'],x['YEAR _5Total assets'],x['YEAR _4Total assets'],x['YEAR _3Total assets'],x['YEAR _2Total assets'],x['YEAR _1Total assets']],axis=1)
dep = pd.concat([x['YEAR _7Depreciation / Amortisation (net of transfer from revaluation reserves)'],x['YEAR _6Depreciation / Amortisation (net of transfer from revaluation reserves)'],x['YEAR _5Depreciation / Amortisation (net of transfer from revaluation reserves)'],x['YEAR _4Depreciation / Amortisation (net of transfer from revaluation reserves)'],x['YEAR _3Depreciation / Amortisation (net of transfer from revaluation reserves)'],x['YEAR _2Depreciation / Amortisation (net of transfer from revaluation reserves)'],x['YEAR _1Depreciation / Amortisation (net of transfer from revaluation reserves)']],axis=1)
volumes_traded=pd.concat([x['L - 7Shares traded'],x['L - 6Shares traded'],x['L - 5Shares traded'],x['L - 4Shares traded'],x['L - 3Shares traded'],x['L - 2Shares traded'],x['L - 1Shares traded']],axis=1)
returns=pd.concat([x['Closing Price L - 7'],x['Closing Price L - 6'],x['Closing Price L - 5'],x['Closing Price L - 4'],x['Closing Price L - 3'],x['Closing Price L - 2'],x['Closing Price L - 1']],axis=1)
returns
```

```python
returns1=returns
```

```python
diff_income=income.diff(axis=1)
```

```python
diff_current_asset=current_assets.diff(axis=1)
diff_current_asset.drop(['YEAR _7Current assets (incl. short term investments, loans & advances)'],inplace=True,axis=1)
diff_cash_balance=cash_balance.diff(axis=1)
diff_cash_balance.drop(['YEAR _7Cash balance'],inplace=True,axis=1)
diff_current_liabilities=current_liabilities.diff(axis=1)
diff_current_liabilities.drop(['YEAR _7Current liabilities'],inplace=True,axis=1)
diff_shortterm_brrw=shortterm_brrw.diff(axis=1)
diff_shortterm_brrw.drop(['YEAR _7Short-term borrowings'],inplace=True,axis=1)
diff_tot_assets=tot_assets.diff(axis=1)
diff_tot_assets.drop(['YEAR _7Total assets'],inplace=True,axis=1)
dep.drop(['YEAR _7Depreciation / Amortisation (net of transfer from revaluation reserves)'],inplace=True,axis=1)
volumes_traded.drop(['L - 7Shares traded'],axis=1,inplace=True)
returns=returns.diff(axis=1)
returns.drop(['Closing Price L - 7'],inplace=True,axis=1)
returns
```

```python
returns
```

```python
returns['returns6']=returns['Closing Price L - 6']/returns1['Closing Price L - 7']
returns['returns5']=returns['Closing Price L - 5']/returns1['Closing Price L - 6']
returns['returns4']=returns['Closing Price L - 4']/returns1['Closing Price L - 5']
returns['returns3']=returns['Closing Price L - 3']/returns1['Closing Price L - 4']
returns['returns2']=returns['Closing Price L - 2']/returns1['Closing Price L - 3']
returns['returns1']=returns['Closing Price L - 1']/returns1['Closing Price L - 2']
returns.drop(['Closing Price L - 6','Closing Price L - 5','Closing Price L - 4','Closing Price L - 3','Closing Price L - 2','Closing Price L - 1'],inplace=True,axis=1)
```

```python
returns
```

```python
accruals6=pd.concat([diff_current_asset['YEAR _6Current assets (incl. short term investments, loans & advances)'],diff_cash_balance['YEAR _6Cash balance'],diff_current_liabilities['YEAR _6Current liabilities'],diff_shortterm_brrw['YEAR _6Short-term borrowings'],dep['YEAR _6Depreciation / Amortisation (net of transfer from revaluation reserves)'],tot_assets['YEAR _6Total assets'],volumes_traded['L - 6Shares traded']],axis=1)
accruals5=pd.concat([diff_current_asset['YEAR _5Current assets (incl. short term investments, loans & advances)'],diff_cash_balance['YEAR _5Cash balance'],diff_current_liabilities['YEAR _5Current liabilities'],diff_shortterm_brrw['YEAR _5Short-term borrowings'],dep['YEAR _5Depreciation / Amortisation (net of transfer from revaluation reserves)'],tot_assets['YEAR _5Total assets'],volumes_traded['L - 5Shares traded']],axis=1)
accruals4=pd.concat([diff_current_asset['YEAR _4Current assets (incl. short term investments, loans & advances)'],diff_cash_balance['YEAR _4Cash balance'],diff_current_liabilities['YEAR _4Current liabilities'],diff_shortterm_brrw['YEAR _4Short-term borrowings'],dep['YEAR _4Depreciation / Amortisation (net of transfer from revaluation reserves)'],tot_assets['YEAR _4Total assets'],volumes_traded['L - 4Shares traded']],axis=1)
accruals3=pd.concat([diff_current_asset['YEAR _3Current assets (incl. short term investments, loans & advances)'],diff_cash_balance['YEAR _3Cash balance'],diff_current_liabilities['YEAR _3Current liabilities'],diff_shortterm_brrw['YEAR _3Short-term borrowings'],dep['YEAR _3Depreciation / Amortisation (net of transfer from revaluation reserves)'],tot_assets['YEAR _3Total assets'],volumes_traded['L - 3Shares traded']],axis=1)
accruals2=pd.concat([diff_current_asset['YEAR _2Current assets (incl. short term investments, loans & advances)'],diff_cash_balance['YEAR _2Cash balance'],diff_current_liabilities['YEAR _2Current liabilities'],diff_shortterm_brrw['YEAR _2Short-term borrowings'],dep['YEAR _2Depreciation / Amortisation (net of transfer from revaluation reserves)'],tot_assets['YEAR _2Total assets'],volumes_traded['L - 2Shares traded']],axis=1)
accruals1=pd.concat([diff_current_asset['YEAR _1Current assets (incl. short term investments, loans & advances)'],diff_cash_balance['YEAR _1Cash balance'],diff_current_liabilities['YEAR _1Current liabilities'],diff_shortterm_brrw['YEAR _1Short-term borrowings'],dep['YEAR _1Depreciation / Amortisation (net of transfer from revaluation reserves)'],tot_assets['YEAR _1Total assets'],volumes_traded['L - 1Shares traded']],axis=1)
```

```python
accruals6['accruals6']=2*((accruals6['YEAR _6Current assets (incl. short term investments, loans & advances)']-accruals6['YEAR _6Cash balance'])-(accruals6['YEAR _6Current liabilities']-accruals6['YEAR _6Short-term borrowings'])-accruals6['YEAR _6Depreciation / Amortisation (net of transfer from revaluation reserves)'])/(accruals6['YEAR _6Total assets']+tot_assets['YEAR _7Total assets'])
accruals5['accruals5']=2*((accruals5['YEAR _5Current assets (incl. short term investments, loans & advances)']-accruals5['YEAR _5Cash balance'])-(accruals5['YEAR _5Current liabilities']-accruals5['YEAR _5Short-term borrowings'])-accruals5['YEAR _5Depreciation / Amortisation (net of transfer from revaluation reserves)'])/(accruals5['YEAR _5Total assets']+tot_assets['YEAR _6Total assets'])
accruals4['accruals4']=2*((accruals4['YEAR _4Current assets (incl. short term investments, loans & advances)']-accruals4['YEAR _4Cash balance'])-(accruals4['YEAR _4Current liabilities']-accruals4['YEAR _4Short-term borrowings'])-accruals4['YEAR _4Depreciation / Amortisation (net of transfer from revaluation reserves)'])/(accruals4['YEAR _4Total assets']+tot_assets['YEAR _5Total assets'])
accruals3['accruals3']=2*((accruals3['YEAR _3Current assets (incl. short term investments, loans & advances)']-accruals3['YEAR _3Cash balance'])-(accruals3['YEAR _3Current liabilities']-accruals3['YEAR _3Short-term borrowings'])-accruals3['YEAR _3Depreciation / Amortisation (net of transfer from revaluation reserves)'])/(accruals3['YEAR _3Total assets']+tot_assets['YEAR _4Total assets'])
accruals2['accruals2']=2*((accruals2['YEAR _2Current assets (incl. short term investments, loans & advances)']-accruals2['YEAR _2Cash balance'])-(accruals2['YEAR _2Current liabilities']-accruals2['YEAR _2Short-term borrowings'])-accruals2['YEAR _2Depreciation / Amortisation (net of transfer from revaluation reserves)'])/(accruals2['YEAR _2Total assets']+tot_assets['YEAR _3Total assets'])
accruals1['accruals1']=2*((accruals1['YEAR _1Current assets (incl. short term investments, loans & advances)']-accruals1['YEAR _1Cash balance'])-(accruals1['YEAR _1Current liabilities']-accruals1['YEAR _1Short-term borrowings'])-accruals1['YEAR _1Depreciation / Amortisation (net of transfer from revaluation reserves)'])/(accruals1['YEAR _1Total assets']+tot_assets['YEAR _2Total assets'])
```

```python
accruals_final=pd.concat([accruals6['accruals6'],accruals5['accruals5'],accruals4['accruals4'],accruals3['accruals3'],accruals2['accruals2'],accruals1['accruals1']],axis=1)
```

```python
accrualsfin=accruals_final.mean(axis=1)
```
